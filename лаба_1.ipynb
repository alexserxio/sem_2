```python
from matplotlib import pyplot as plt
import numpy as np
from scipy import optimize


def func(x):  # левая часть уравнения
    return x**3 + x**2 - 22.25 * x - 11.25


eps = 1e-6
# метод итераций
def iter(x0, coeff):
    global eps
    x = x0 + 1
    while abs(x0 - x) > eps:
        x0 = x
        x = x0 - coeff * func(x0)
    return x


# создаем массив от -10 до 10
x = np.arange(-10, 10, 0.1)
y = func(x)  # вычисляем значения y
plt.plot(x, y)  # рисуем график
plt.xlabel("x")
plt.ylabel("y")
plt.grid()  # включаем сетку
plt.show()  # выводим график на экран


# Выводим наши значения и scipy
print("y равен 0 при x равном:\nметод итер scipy")
print(
    round(iter(-5.5, 0.01), 6),
    optimize.bisect(func, -5.5, -4.5, xtol=eps, maxiter=100),
)
print(
    round(iter(1, -0.01), 6),
    optimize.bisect(func, -1, 1, xtol=eps, maxiter=100),
)
print(
    round(iter(4, 0.01), 6),
    optimize.bisect(func, 4, 5, xtol=eps, maxiter=100),
)
```

```python
from matplotlib import pyplot as plt
import numpy as np


# Правая часть уравнение
def fun(E, r, I):
    return I - E / (r + 10 * np.log(I + 1))


E = 100
r = 20
Imax=E/r
# Создание массивов координат точек графика
I = np.arange(0, Imax, 0.01)
Y = fun(E, r, I)
# Построение графика
plt.plot(I, Y)
plt.grid()
plt.show()
```

```python
from scipy.optimize import bisect
import numpy as np


# Правая часть уравнение
def fun(I, E, r):
    return I - E / (r + 10 * np.log(I + 1))


E = 100
r = 20
I = bisect(fun, 2, 3, args=(E, r))
print("Ток схемы равен", round(I, 6))
```

```python
from scipy.interpolate import lagrange
import numpy as np
from matplotlib import pyplot as plt


H = [0, 2500, 5000, 10000]
B = [0, 1.5, 1.62, 1.75]


lagr = lagrange(H, B)  # Получаем полином Лагранжа для наших значений
x = np.arange(0, 10000, 5)
y = lagr(x)  # с помощью полинома вычисляем значения y
plt.plot(x, y, "r")  # рисуем график использованной интерполяции
plt.plot(H, B, "bo")  # рисуем точки данных
plt.grid()
plt.xlabel("H, А/м")
plt.ylabel("B, Тл")
plt.show()
```
